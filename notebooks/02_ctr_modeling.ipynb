# iPinYou CTR Modeling

This notebook prepares modeling features, trains a logistic regression baseline and a stronger gradient boosting model, and compares validation metrics.

```python
from pathlib import Path

from ipinyou_analysis.data import load_ipinyou_logs
from ipinyou_analysis.features import build_modeling_frame
from ipinyou_analysis.modeling import evaluate_models, train_ctr_models

project_root = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
data_dir = project_root / "data" / "raw"

df = load_ipinyou_logs(data_dir)
modeling_df = build_modeling_frame(df)
modeling_df.head()
```

```python
artifacts = train_ctr_models(modeling_df)
metrics = evaluate_models(artifacts)
metrics
```

```python
best_model_name = metrics.iloc[0]["model"]
best_model_name
```
